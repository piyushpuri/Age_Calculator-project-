```python
def Age_Calculator():
    user_input=int(input("What's Your Age? "))
    print(f"")
    if user_input<1:
        print(f"You Have Entered Incorrect Age")
        print(f"")
        try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
        print(f"")
        if try_again=="yes":
            Age_Calculator()
        elif try_again!="no":
            print(f"")
            print(f"You Have Typed Wrong input")
            print(f"")
            
    elif user_input>0:
        print(f"Please Choose Time Units ")
        print(f"")
        print(f"Type:")
        print(f"")
        print(f"'m' for Month")
        print(f"'w' for Week")
        print(f"'d' for Days")
        print(f"'h' for Hours")
        print(f"'minutes' for Minutes")
        print(f"'s' for Seconds")
        user_input1=input(" ")
        if user_input1=="m" or user_input1=="w" or user_input1=="d" or user_input1=="h" or user_input1=="minutes" or user_input1=="s":
            
            if user_input1=="m":
                user_input1=user_input*12
                print(f"")
                print(f"You Lived For {user_input1} Months")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
                elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
                elif try_again!="no":
                    print(f"")
                    print(f"You Have Typed Wrong input")
                    print(f"")
                
            elif user_input1=="w":
                
                user_input1=user_input*52
                print(f"")
                print(f"You Lived For {user_input1} Weeks")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
                elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
                elif try_again!="no":
                    print(f"")
                    print(f"You Have Typed Wrong input")
                    print(f"")
                    
            elif user_input1=="d":
                
                user_input1=user_input*365
                print(f"")
                print(f"You Lived For {user_input1} Days")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
                elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
                elif try_again!="no":
                    print(f"")
                    print(f"You Have Typed Wrong input")
                    print(f"")
                    
            elif user_input1=="h":
                user_input1=user_input*24*365
                print(f"You Lived For {user_input1} Hours")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
                elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
                elif try_again!="no":
                    print(f"")
                    print(f"You Have Typed Wrong input")
                    print(f"")
                
            elif user_input1=="minutes":
                user_input1=user_input*60*24*365
                print(f"")
                print(f"You Lived For {user_input1} Minutes")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
                elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
                elif try_again!="no":
                    print(f"")
                    print(f"You Have Typed Wrong input")
                    print(f"")
                
            elif user_input1=="s":
                user_input1=user_input*60*60*24*365
                print(f"")
                print(f"You Lived For {user_input1} Seconds")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
                elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
                elif try_again!="no":
                    print(f"")
                    print(f"You Have Typed Wrong input")
                    print(f"")
            
        else:
            print(f"You Have Type Incorrect Option")
            print(f"")
            try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
            print(f"")
            if try_again=="yes":
                Age_Calculator()
            elif try_again=="no":
                    print(f"Thank You For Using Age Calculator")
            elif try_again!="no":
                print(f"")
                print(f"You Have Typed Wrong input")
                print(f"")
                try_again=input("Do You Want To Try Again Please Type 'yes' or 'no': ")
                print(f"")
                if try_again=="yes":
                    Age_Calculator()
        
              
Age_Calculator()
```
